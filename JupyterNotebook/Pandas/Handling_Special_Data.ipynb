# Common methods to replace values using replace() 

```python
import pandas as pd
df = pd.read_csv("F:\\Skill Develpmn\\CodeBasics\\weather_data1.csv")
df
```

## Using replace() method to substitute the reqd values with NaN

```python
import numpy as np
df1 = df.replace([-99999,-88888],np.NaN)
df1
```

```python
df2 = df1.replace({
    'event' : ['0', 'No event']
}, np.NaN)
df2
```

Changing table data now in order to make it messy for data cleaning using regex.

```python
df2.iloc[0,2] = "6.0 mph"
df2.iloc[1,2] = "7.0 MilesPerHour"
df2.iloc[0,1] = '32.0 C'
df2.iloc[2,1] = '28.0 degreeC'
df2
```

**Using regex function, Removing unwanted characters from columns so that only numeric value is retained**

```python
df3 = df2.replace({
    'day' : '[A-Za-z]',
    'temperature' : '[A-Za-z]',
    'windspeed' : '[A-Za-z]'
}, '',  regex =True)
df3
```

numbers are desired in DS and not letters and other characters in the DF. So whenever possible we should replace them with numbers. Lets say in above table, in event column, replace sunny==>1, rain==>2 and snow ==>3 ()

```python
df4 = df3.replace(['Sunny', 'Rain', 'Snow'],[1,2,3])   #replacing a list of values with other list
df4
```
