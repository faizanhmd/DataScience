Suppose there are "not available" or "NA" values in ur CSV file which is not the case in this particular file. (So we will change a missing/NaN cell to "N.A" so that,) We can use na_values method to convert such cell value to NaN.

```python
df.iloc[0,9] = "N.A"   #changes the firstvalue of column Events to "N.A" which was NaN earlier
df.head()
```

```python
df = pd.read_csv("F:\\Skill Develpmn\\CodeBasics\\nyc_weather.csv", na_values = ["N.A"])  #See it finds the "N.A" in df and changes it to NaN to be dealt later.
df.head()
```

Sometimes there are wrong data also present which should be converted to NaN. So We can also use dictionary in na_values method so that different columns have different criterion to be searched and then put as NaN. 

```python
df = pd.read_csv("F:\\Skill Develpmn\\CodeBasics\\nyc_weather.csv", na_values = {
    "EST" : ["N.A","n.a.","N.A.","n.a"],
    "Temperature " : ["N.A","n.a.","N.A.","n.a"],
    "Dewpoint" : ["N.A","n.a."],
    "WindSpeedMPH" : ["N.A","n.a.", -1]   })
df.head()
```

Writing an excel File with two sheets: use ExcelWriter() class

```python
Stocks = {
    "tikcers": ['Reliance','BSNL','Tata'],
    "price" : [6200,460,1600],
    "eps" : [27.8,3.7,11.5]
}

Inventory = {
    "Company" : ['Parle', 'Britannia', 'Nestle', 'Cadbury'],
    "Product" : ['parleG','50-50', 'Maggi', 'Dairy Milk'],
    "Quantity" : [40,30,60,54]
}

df1 = pd.DataFrame(Stocks)
df2 = pd.DataFrame(Inventory)
df2
```

```python
with pd.ExcelWriter('F:\\Skill Develpmn\\CodeBasics\\SampleExcel.xlsx') as writer:
    df1.to_excel(writer, sheet_name = 'stocks', index = False)
    df2.to_excel(writer, sheet_name = 'inventory', index = False) 
```
