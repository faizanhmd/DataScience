# Methods for concatenation of df and Series

```python
import pandas as pd
df1 = pd.read_csv('F:\\Skill Develpmn\\CodeBasics\\weather_data.csv')
df1
```

```python
df2 = pd.read_csv('F:\\Skill Develpmn\\CodeBasics\\weather_data1.csv')
df2
```

```python
# NOw to concat these two above df in single df:
df3 = pd.concat([df1, df2])
df3
```

Note that index restarts from 0 for the second df, so in order to continue the index:

```python
df3 = pd.concat([df1, df2], ignore_index = True)
df3
```

```python
df3 = pd.concat([df1, df2], axis = 1, ignore_index =True)
df3
```

### Concatenation of df by using dictionaries

```python
df1  = pd.DataFrame({
    "cities" : ['mumbai', 'new delhi', 'kolkata'],
    "humidity" : [80, 65, 75],
    "temperature" : [30, 42, 34]
})
df2 = pd.DataFrame({
    "cities" : ['new york', 'chicago', 'la'],
    "humidity" : [72, 76, 68],
    "temperature" : [36, 28, 33]
})
df3 = pd.concat([df1, df2], ignore_index = True)
df3
```

```python
df4 = pd.concat([df1, df2], keys = ['ind', 'US'])  #Use of keys
df4
```

```python
df4.loc['ind']
```

```python
df4.loc['US']
```

```python
wsdf = pd.DataFrame({
    "cities" : ['mumbai', 'kolkata', 'new delhi'],
    "windspeed" : [9,8,6] 
}, index = [0,2,1])
wsdf
```

```python
df5 = pd.concat([df1, wsdf], axis = 1)
df5
```

To concat df with series:

```python
s = pd.Series(['humid','sunny','rain'], name = 'event')
s
```

```python
df6 = pd.concat([df1,s], axis = 1)
df6
```
