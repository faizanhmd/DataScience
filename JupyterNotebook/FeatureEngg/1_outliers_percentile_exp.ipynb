Notice that price(per night's min value is 0{free} and max is 10,000{very high}). We can print these values and see if these are errors and must be removed as OUTLIERS.

```python
#For all rows having price = 0
df.loc[df['price']==0]
```

```python
#For all rows having price = 10000
df.loc[df['price']==10000]
```

We can use Percentile method for Outlier removal since we dont have the domain expertise to obtain the exact price per night of rooms in New York city. Lets see the range after having a percentile of 99.5% and 0.1% for max and min threshold.

```python
max_thresh = df.price.quantile(0.995)
max_thresh
```

```python
min_thresh = df.price.quantile(0.001)
min_thresh
```

```python
df1 = df[(df.price < max_thresh) & (df.price > min_thresh)]
```
