```python
import pandas as pd
df = pd.read_csv("F:\\Skill Develpmn\\CodeBasics\\weather_data.csv", parse_dates = ["day"])
df
```

```python
type(df.day[0])   #checking how the data type changed
```

```python
#Setting day column as index using set_index
df.set_index('day',inplace = True)
df.head()
```

## 1. Using fillna() method to replace NaN values with the desired value.

```python
df1 = df.fillna(0)
df1
```

```python
#Use dictionary if different values to be replaced for different columns(Like 0 for Event do not makes sense.)
df1 = df.fillna({
    "temperature" : 30,
    "windspeed" : 7,
    "event" : "No event"
})
df1
```

```python
#Instead of using random numbers in place of NaN; if u simply want to carry forward the previous value use method='ffill'
df2 = df.fillna(method = 'ffill')
df2
```

```python
#Similarly for backfilling, we have:
df3 = df.fillna(method = 'bfill')
df3
```

```python
#Using 'limit' to limit the cells upto which the filling is applied
df4 = df.fillna(method = 'ffill', limit =1)
df4
```

## 2. Using interpolate() method to replace NaN with desired values.

```python
df5 = df.interpolate()  #No arguement means default method which is linear interpolation
df5
```

```python
df5 = df.interpolate(method = "quadratic")  #For cubic interpolation
df5
```

```python
df5 = df.interpolate(method = 'time')  #It calculates according to time i.e Date or TiemStamp
df5
```

##  3. Using dropna() method to remove the rows which has NaN.

```python
df6 = df.dropna()  #drops all the rows which has even a single NaN
df6
```

```python
df6 = df.dropna(how = 'all')  #drops all the rows which only has all NaN values
df6
```

```python
df6 = df.dropna(thresh = 2)  #drops all the rows which do not have at least 2 non NaN values.
df6
```
