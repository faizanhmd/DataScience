# To predict home prices using ML technique 'Simple Linear Regression'

Below table represents current home prices in monroe township based on square feet area, new jersey

**Problem Statement:** To build a machine learning model that can predict home prices based on square feet area. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("F:\\Skill Develpmn\\CodeBasics\\homeprices.csv")
df
```

first of all draw scatter plot in order to have visual understanding of the data (So that it will be evident to use linear regression model in such a case)

```python
%matplotlib inline
plt.xlabel('area(sq feet)')
plt.ylabel('price($)')
plt.scatter(df.area, df.price, color = 'red', marker = '+')
```

A linear equation is to be established between the prices or y (dependant variable) and area or x (indep variable), so that a straight line can be used as a pattern for prediction of any new value. 
Error(delta) is the offset of the exact points from this straight line. The sum of the square of these individual errors should be minimum for the striaght line being selected to depict the linear relation.

So simply we have now converted the problem to:

```python
reg = linear_model.LinearRegression()    # Create linear regression object(reg)
reg.fit(df[['area']], df.price)   #fit is used for training the model. Also the 1st arg in fit() need to be a 2D array
```

```python
reg.predict([[3300]])
```

What model did behind the scene is to calculate the coefficient(or gradient or slope) and intercept based on the training data, and then use them while predicting any new value of x.

```python
reg.coef_
```

```python
reg.intercept_
```

Y = m * X + b (m is coefficient and b is intercept)

```python
135.78767123*3300+180616.43835616432  #will be equal to the value predicted by the model above.
```

To see how linear equation line look like:

```python
%matplotlib inline
plt.xlabel('area(sq feet)')
plt.ylabel('price($)')
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.plot(df.area, reg.predict(df[['area']]), color = "blue")
```
